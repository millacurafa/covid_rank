```python
import pandas as pd
import plotly.express as px
from raceplotly.plots import barplot
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/lc5415/raceplotly/main/example/FAOSTAT_data.csv')

```

```python
data.head()
```

```python
##Imports regional data
df_city_current = pd.read_csv("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto25/CasosActualesPorComuna_std.csv")
```

```python
# regional daily active cases

df_region_current = df_city_current[df_city_current["Comuna"] == "Total"]

df_region_current_bypop = df_city_current[df_city_current["Comuna"] == "Total"]
df_region_current_bypop['bypop'] = (df_region_current.loc[0:,"Casos actuales"]/(df_region_current.loc[0:,"Poblacion"]/10000)).round(1)
df_region_current_bypop = df_region_current_bypop[["Region", "Fecha", "bypop", 'Casos actuales']]
df_region_current_bypop
```

```python
df_region_current_bypop['Region'].unique()
```

```python
df_region_current_bypop.info()
```

```python
df_region_current_bypop['Fecha'] = pd.to_datetime(df_region_current_bypop['Fecha'])
```

```python
df_region_current_bypop['Region'] = df_region_current_bypop['Region'].str.replace('Del Libertador General Bernardo O’Higgins','O’Higgins')
```

```python
colors = {'Arica y Parinacota': 'rgb(255, 214, 165)',
          'Tarapaca': 'rgb(255, 198, 255)',
          'Antofagasta': 'rgb(253, 255, 182)',
          'Atacama':'rgb(208, 210, 211)',
          'Coquimbo':'rgb(208, 210, 211)',
          'Valparaiso':'rgba(40, 95, 127, 1)',
          'Metropolitana':'rgb(208, 210, 211)',
          'O’Higgins':'rgb(208, 210, 211)',
          'Maule':'rgb(208, 210, 211)',
          'Nuble':'rgb(208, 210, 211)',
          'Biobio':'rgb(189, 178, 255)',
          'La Araucania':'rgb(202, 255, 191)',
          'Los Rios':'rgb(155, 246, 255)',
          'Los Lagos':'rgb(160, 196, 255)',
          'Aysen':'rgb(208, 210, 211)',
          'Magallanes y la Antartica':'rgb(255, 173, 173)'
         }
```

```python
df_region_current_bypop['color'] = df_region_current_bypop['Region'].map(colors)
```

```python
covid_race = barplot(df_region_current_bypop,
                     item_column='Region',
                      value_column='bypop',
                      time_column='Fecha',
                     top_entries = 16,
                     item_color='colors'
)
```

```python
import locale
locale.setlocale(locale.LC_TIME, 'es_CL')
```

```python
covid_race.plot(title = 'Casos activos de Covid19 por region',
                 item_label = 'Region',
                 value_label = 'Casos activos por cada 10000 habitantes',
                 frame_duration = 500,
                 time_label = '',
                 date_format = '%d de %B %Y',
                 ).update_layout(font=dict(size=18),
                                    height=700,
                                )
```
